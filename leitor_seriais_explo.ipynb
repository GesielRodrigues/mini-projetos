# Importando Bibliotecas

```python
import xml.etree.ElementTree as ET 
from google.colab import files 
```

# Lendo o Arquivo XML

```python
# Upload do arquivo
upload = files.upload()
```

```python
# Pegando o nome do arquivo
xml = list(upload.keys())[0]
```

```python
# Lendo o XML e pegando a raiz
tree = ET.parse(xml)
root = tree.getroot()
```

# Funções

```python
def getNomeProdutos(root):
  nomeProdutos = []
  for child in root:
    nomeProdutos.append(child.attrib['nome'])
  return nomeProdutos
```

```python
def getAnfo():
  produtos = root.findall('{http://www.dfpc.eb.mil.br}produto')
  for p in produtos:
    if p.attrib['nome'].startswith("ANFOMAX"):
      itens = p.findall('{http://www.dfpc.eb.mil.br}itens')
  for i in itens:
    iis = i.findall('{http://www.dfpc.eb.mil.br}item')
    print('GRANULADO')
    for serial in iis:
      print(serial.attrib['iis'])



def getEncartuchado():
  produtos = root.findall('{http://www.dfpc.eb.mil.br}produto')
  for p in produtos:
    if p.attrib['nome'].startswith("IBEGEL"):
      itens = p.findall('{http://www.dfpc.eb.mil.br}itens')
  for i in itens:
    iis = i.findall('{http://www.dfpc.eb.mil.br}item')
    print('DINAMITE')
    for serial in iis:
      print(serial.attrib['iis'])


def getCordelCx(np='10'):
  produtos = root.findall('{http://www.dfpc.eb.mil.br}produto')
  for p in produtos:
    if p.attrib['nome'].startswith(f"BRITACORD NP {np}"):
      itens = p.findall('{http://www.dfpc.eb.mil.br}itens')
  for i in itens:
    iis = i.findall('{http://www.dfpc.eb.mil.br}item')
    print(f'CORDEL NP {np}')
    for serial in iis:
      print(serial.attrib['iis'])


def getCordel(np='10'):
  produtos = root.findall('{http://www.dfpc.eb.mil.br}produto')
  for p in produtos:
    if p.attrib['nome'].startswith(f"BRITACORD NP {np}"):
      itens = p.findall('{http://www.dfpc.eb.mil.br}itens')
  for i in itens:
    iis = i.findall('{http://www.dfpc.eb.mil.br}item')
    print(f'CORDEL NP {np}')
    for serial in iis:
     for s in serial:
        print(s.attrib['iis'])

def getRetardos(ms=30):
  produtos = root.findall('{http://www.dfpc.eb.mil.br}produto')
  for p in produtos:
    if p.attrib['nome'].startswith(f"BRITACRON {ms}"):
      itens = p.findall('{http://www.dfpc.eb.mil.br}itens')
  for i in itens:
    iis = i.findall('{http://www.dfpc.eb.mil.br}item')
    print(f"RETARDO {ms}")
    for serial in iis:
      for s in serial:
        print(s.attrib['iis'])


def getRetardosSimples(ms=30):
  produtos = root.findall('{http://www.dfpc.eb.mil.br}produto')
  for p in produtos:
    if p.attrib['nome'].startswith(f"BRITACRON {ms}"):
      itens = p.findall('{http://www.dfpc.eb.mil.br}itens')
  for i in itens:
    iis = i.findall('{http://www.dfpc.eb.mil.br}item')
    print(f"RETARDO {ms}")
    for serial in iis:
      print(serial.attrib['iis'])

def getRetardosItens(ms=30):
  produtos = root.findall('{http://www.dfpc.eb.mil.br}produto')
  for p in produtos:
    if p.attrib['nome'].startswith(f"BRITACRON {ms}"):
      itens = p.findall('{http://www.dfpc.eb.mil.br}itens')
  for i in itens:
    iis = i.findall('{http://www.dfpc.eb.mil.br}item')
    print(f"RETARDO {ms}")
    for serial in iis:
      ss = serial.findall('{http://www.dfpc.eb.mil.br}item')
      for ii in ss:
        s = ii.findall('{http://www.dfpc.eb.mil.br}item')
        for retardo in s:
          print(retardo.attrib['iis'])
```
