## 1 - Importar as bibliotecas.

```python
import pandas as pd
import time

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
```

## 2 - Iniciar o navegador (chrome) e abrir o site etf.com

```python
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service = service)

url = "https://www.etf.com/etfanalytics/etf-finder"

driver.get(url)
```

## 3 - Encontrar os elementos necessários dentro do HTML do site

```python
# 10s de delay para dar tempo da tabela carregar por completo
time.sleep(5)
```

```python
#  Expandindo a tabela para 100 linhas

botao_100 = driver.find_element(By.XPATH, '/html/body/div[4]/section/div/div[3]/section/div/div/div/div/div[2]/section[2]/div[2]/section[2]/div[1]/div/div[4]/button/label/span')

driver.execute_script("arguments[0].click();", botao_100)
```

```python
# Pegar o número de páginas da tabela

numero_paginas = driver.find_element(By.ID, 'totalPages')

numero_paginas = numero_paginas.text.replace("of ", "")

numero_paginas = int(numero_paginas)

print(numero_paginas)
```

## 4 - Ler a tabela de dados

### 4.1 - Tabela Fund Basics

```python
# Lista para armazenar as tabelas por página
lista_tabelas = []

for pagina in range(0, numero_paginas):
    # Lendo elemento da tabela
    tabela = driver.find_element(By.ID, 'finderTable')
    # Extraindo o HTML do elemento tabela
    html_tabela = tabela.get_attribute("outerHTML")
    # Transformando o HTML em um dataframe do pandas
    df = pd.read_html(html_tabela)[0]
    # Adicionando o dataframe a lsita de tabelas por página
    lista_tabelas.append(df)
    # Clicar no botão de próxima página
    botao_avancar_pagina = driver.find_element(By.ID, 'nextPage')
    driver.execute_script("arguments[0].click();", botao_avancar_pagina)
    
# Concatenando as tabelas de cada uma das páginas em um dataframe apenas
dados_fund_basics = pd.concat(lista_tabelas)

display(dados_fund_basics)
```

### 4.2 - Tabela Performance

```python
# Trocar para a aba de Performance
aba_performance = driver.find_element(By.XPATH, '//*[@id="table-tabs"]/li[2]/span')
driver.execute_script("arguments[0].click();", aba_performance)

# Voltando para a primeira página da tabela
num_paginas = driver.find_element(By.ID, 'goToPage')
num_paginas.clear()
num_paginas.send_keys('1')
num_paginas.send_keys(Keys.ENTER)
```

```python
# Lista para armazenar as tabelas por página
lista_tabelas = []

for pagina in range(0, numero_paginas):
    # Lendo elemento da tabela
    tabela = driver.find_element(By.ID, 'finderTable')
    # Extraindo o HTML do elemento tabela
    html_tabela = tabela.get_attribute("outerHTML")
    # Transformando o HTML em um dataframe do pandas
    df = pd.read_html(html_tabela)[0]
    # Adicionando o dataframe a lsita de tabelas por página
    lista_tabelas.append(df)
    # Clicar no botão de próxima página
    botao_avancar_pagina = driver.find_element(By.ID, 'nextPage')
    driver.execute_script("arguments[0].click();", botao_avancar_pagina)
    
# Concatenando as tabelas de cada uma das páginas em um dataframe apenas
dados_performance = pd.concat(lista_tabelas)

display(dados_performance)
```

```python
# Fechar o navegador
driver.quit()
```

## 5 - Construindo a tabela final

```python
# Alterando os índices da tabela para os Tickers dos ETF's
dados_fund_basics = dados_fund_basics.set_index("Ticker")
dados_performance = dados_performance.set_index("Ticker")

# Pegando da tabela Performance apenas os dados de 1, 5 e 10 anos
dados_performance = dados_performance[['1 Year', '5 Years', '10 Years']]

# Juntando os dados
base_de_dados_final = dados_fund_basics.join(dados_performance)
display(base_de_dados_final)
```

## 6 - Exportar para csv

```python
base_de_dados_final.to_csv('files\etfs.csv')
```
